```python
#KATA 1

SELECT
  --EXTRACT(DAY FROM events.created_at) as day,
  CAST(events.created_at AS DATE) as day,
  events.description as description,
  COUNT(events.created_at) as count
FROM
  events
WHERE 
  name = 'trained'
GROUP BY 
  day, description
ORDER BY 
  day;
```

```python
#KATA 2

SELECT
  capital
FROM
  countries 
WHERE 
  (continent = 'Africa' OR continent = 'Afrika')
  AND
  (country LIKE 'E%')
ORDER BY
  capital
LIMIT 3;
```

```python
#KATA 3

SELECT 
  CAST(s.transaction_date AS DATE) AS day,
  d.name AS department,
  COUNT(*) AS sale_count
FROM 
  department d
JOIN 
  sale s ON d.id = s.department_id
GROUP BY 
  day, department
ORDER BY 
  day ASC, department ASC;
```
